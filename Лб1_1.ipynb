```python
import numpy as np
import matplotlib.pyplot as plt

#Генерація вибірки з 1000 елементів з рівномірним розподілом від -1 до 7
data = np.random.uniform(low=-1, high=7, size=1000)

#Побудова гістограми
plt.hist(data, bins=30, alpha=0.7, histtype='stepfilled', color='steelblue', edgecolor='none')
plt.title('Гістограма рівномірного розподілу')
plt.xlabel('Значення')
plt.ylabel('Частота')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Визначення функцій
def f1(x):
    return x**2

def f2(x):
    return 10 * np.sin(x)

# Створення масиву значень x від -5 до 12
x_values = np.linspace(-5, 12, 1000)

#Обчислення значень функцій
y1_values = f1(x_values)
y2_values = f2(x_values)
y_sum = y1_values + y2_values

#Побудова графіків
plt.figure(figsize=(10, 6))
plt.plot(x_values, y1_values, label='$f(x) = x^2$', color='blue')
plt.plot(x_values, y2_values, label='$f(x) = 10\sin(x)$', color='orange')
plt.plot(x_values, y_sum, label='$f(x) = x^2 + 10\sin(x)$', color='green')

#Відображення графіка
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Параметри сфери
R = 1
theta = np.linspace(0, np.pi / 2, 100)
phi = np.linspace(0, 2 * np.pi, 100)

#Параметри верхньої половини сфери
theta, phi = np.meshgrid(theta, phi)
x = R * np.sin(theta) * np.cos(phi)
y = R * np.sin(theta) * np.sin(phi)
z = R * np.cos(theta)

#Побудова 3D графіку
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

#Побудова верхньох половини сфери
ax.plot_surface(x, y, z, color='gray', alpha=0.8)

#Налаштування графіку
ax.set_title('Upper Half of a Sphere')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Створення графіку
x = np.linspace(-10, 10, 1000)
y = np.sqrt(100 - x**2)

plt.plot(x, -y, color='red')
plt.plot([-10, -10], [10, -y[np.where(x==-10)[0][0]]], linestyle='--', color='orange')
plt.plot([10, 10], [10, -y[np.where(x==10)[0][0]]], linestyle='--', dashes=(0.5, 1), color='green')
plt.plot([-10, 10], [10, 10], color='blue')

#Налаштування осей
plt.xlim(-16, 16)
plt.ylim(-11, 11)
plt.xticks(np.arange(-15, 16, step=5))
plt.yticks(np.arange(-10, 11, step=2.5))
plt.xlabel('X')
plt.ylabel('Y')

#Відображення графіку
plt.show()
```
