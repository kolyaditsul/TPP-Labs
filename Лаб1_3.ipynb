```python
import pandas as pd

data = pd.read_csv('math_students.csv', delimiter=',')
df = pd.DataFrame(data)
print("Исходная таблица:")
print(df.head())  # Выводим первые несколько строк для изучения структуры данных

#1.Получите таблицу со студентами, имеющими четное число пропусков.
students_even_indices = df[df.index % 2 == 0]
print("Таблица со студентами с четными номерами:")
print(students_even_indices)

#2.Получите таблицу со студентами, имеющими четное число пропусков.
students_even_absences = df[df['absences'] % 2 == 0]
print("Таблица со студентами, имеющими четное число пропусков:")
print(students_even_absences)

# 3. Верно ли, что студенты, имеющие больше 10 пропусков, учатся хуже, чем остальные?
more_than_10_absences = df[df['absences'] > 10]
less_than_or_equal_10_absences = df[df['absences'] <= 10]
average_grade_more_than_10 = more_than_10_absences['G3'].mean()
average_grade_less_than_or_equal_10 = less_than_or_equal_10_absences['G3'].mean()
print("\nСредняя оценка студентов с более чем 10 пропусками:", average_grade_more_than_10)
print("Средняя оценка студентов с 10 и менее пропусками:", average_grade_less_than_or_equal_10)
print("Вывод: Студенты с более чем 10 пропусками учатся хуже.")

#4.Как вы думаете, какие еще признаки могут повлиять на итоговую оценку? Найдите закономерности, аналогичные рассмотренным.
 # Время на учебу помимо школы
study_time_grouped = df.groupby('studytime')['G3'].mean()
print("\nСредняя оценка в зависимости от времени на учебу помимо школы:")
print(study_time_grouped)

# Наличие доступа к интернету
internet_grouped = df.groupby('internet')['G3'].mean()
print("\nСредняя оценка в зависимости от наличия доступа к интернету:")
print(internet_grouped)

#5.Как вы думаете, если при школе откроется общежитие, насколько актуально это будет для студентов математического класса?
# Отбор индексов студентов математического класса и других студентов.
#Выберем студентов математического класса, предполагая, что они имеют высокие оценки (G1, G2, G3) по математике.
#Предположим также, что студенты с оценками G3 выше среднего являются студентами математического класса. Вот пример кода для этого:
math_student_indices = df[df['G3'] > df['G3'].mean()].index
other_student_indices = df[df['G3'] <= df['G3'].mean()].index

# Расчет среднего времени в пути для студентов математического класса и других студентов
traveltime_math_students = df.loc[math_student_indices, 'traveltime'].mean()
traveltime_other_students = df.loc[other_student_indices, 'traveltime'].mean()

# Вывод результатов
print("\nСреднее время в пути для студентов математического класса:", traveltime_math_students)
print("Среднее время в пути для студентов других классов:", traveltime_other_students)

# Вывод вывода
if traveltime_math_students > traveltime_other_students:
    print("Вывод: Если у студентов математического класса среднее время в пути больше, общежитие может быть актуальным.")
else:
    print("Вывод: Если у студентов математического класса среднее время в пути меньше или равно, общежитие может быть менее актуальным.")

#6.Найдите в данных закономерности с помощью группировки.
# Группировка данных по наличию свободного времени после школы и расчет средней оценки (G3)
freetime_grouped = df.groupby('freetime')['G3'].mean()

# Вывод результатов
print("\nСредняя оценка в зависимости от количества свободного времени после школы:")
print(freetime_grouped)

#По полу или уровню образования родителе
grouped_by_gender = df.groupby('sex')['G3'].mean()
grouped_by_edu = df.groupby('Medu')['G3'].mean()

print("\nСредняя оценка в зависимости от пола:")
print(grouped_by_gender)

print("\nСредняя оценка в зависимости от уровня образования матери:")
print(grouped_by_edu)
```
